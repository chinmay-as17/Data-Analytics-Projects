```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv('Top 10 Products with High Discounts & Low Revenue.csv')

# Plot
plt.figure(figsize=(10,6))
sns.scatterplot(
    x='AvgDiscount',
    y='TotalRevenue',
    data=df,
    hue='ProductID',
    palette='tab10',
    s=100
)

plt.title('Products with High Discount & Low Revenue')
plt.xlabel('Average Discount (%)')
plt.ylabel('Total Revenue')
plt.legend(title='ProductID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
cat_sales = pd.read_csv('Category-wise Sales Summary.csv')

plt.figure(figsize=(8,5))
sns.barplot(x='TotalSales', y='Category', data=cat_sales.sort_values(by='TotalSales', ascending=False), color='skyblue')
plt.title('Category-wise Sales (SQL Result)')
plt.tight_layout()
plt.show()
```

```python
vendor_rank = pd.read_csv('Vendor Ranking from Vendor Summary.csv')

plt.figure(figsize=(10,6))
sns.barplot(x='GP_Percent', y='VendorName', data=vendor_rank.sort_values(by='GP_Percent'), palette='viridis')
plt.title('Vendor Rankings (From SQL Output)')
plt.tight_layout()
plt.show()

```

```python
df = pd.read_csv('Category-wise Sales Summary.csv')

df = df.sort_values(by='TotalSales', ascending=False)
df['Cumulative%'] = df['TotalSales'].cumsum() / df['TotalSales'].sum() * 100

plt.figure(figsize=(10,5))
sns.barplot(x='Category', y='TotalSales', data=df, color='skyblue')
plt.plot(df['Category'], df['Cumulative%'], color='red', marker='o')
plt.axhline(80, color='green', linestyle='--')

plt.title('Pareto Chart: Category-wise Revenue Contribution')
plt.ylabel('Total Sales & Cumulative %')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
